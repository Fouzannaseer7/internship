```python
# Given values
P_D = 0.01               # Probability of disease
P_Pos_D = 0.99           # Test is positive if diseased
P_No_D = 1 - P_D         # No disease
P_Pos_No_D = 0.05        # Test is positive if not diseased

# Total probability of a positive test
P_Pos = P_Pos_D * P_D + P_Pos_No_D * P_No_D

# Bayes' Theorem
P_D_given_Pos = (P_Pos_D * P_D) / P_Pos

print("P(Disease | Positive Test):", round(P_D_given_Pos, 4))
```

```python
Problem 1.1: Coin Toss
```

```python
Q: What's the probability of getting exactly 2 heads in 3 coin tosses?
```

```python
Solution:
```

```python
List all possible outcomes (n=8):
HHH, HHT, HTH, THH, HTT, THT, TTH, TTT

Count favorable outcomes: HHT, HTH, THH ‚Üí 3

Probability = 3/8 = 0.375
```

```python
from itertools import product

outcomes = list(product(['H','T'], repeat=3))
favorable = [toss for toss in outcomes if toss.count('H') == 2]
prob = len(favorable)/len(outcomes)
print(f"Probability: {prob:.3f}")  # 0.375
```

```python
Problem 1.2: Dice Probability
```

```python
Q: What's the probability of rolling a sum of 7 with two dice?
```

```python
Solution:

Total outcomes: 6 √ó 6 = 36

Favorable combinations:
(1,6), (2,5), (3,4), (4,3), (5,2), (6,1) ‚Üí 6

Probability = 6/36 = 1/6 ‚âà 0.1667
```

```python
2. Conditional Probability Problems
```

```python
Problem 2.1: Medical Testing
```

```python
Scenario:

Disease prevalence: 1%
Test accuracy:
 99% true positive rate
 5% false positive rate
```

```python
Q: If someone tests positive, what's the probability they actually have the disease?
```

```python
Solution (Bayes' Theorem):
```

```python
P(Disease|Positive) = [P(Positive|Disease) √ó P(Disease)] / P(Positive)
                   = (0.99 √ó 0.01) / [(0.99√ó0.01) + (0.05√ó0.99)]
                   ‚âà 0.1667 or 16.67%
```

```python
Python Calculation:
```

```python
p_disease = 0.01
p_pos_given_disease = 0.99
p_pos_given_no_disease = 0.05

numerator = p_pos_given_disease * p_disease
denominator = numerator + (p_pos_given_no_disease * (1 - p_disease))
result = numerator / denominator
print(f"Probability: {result:.4f}")  # 0.1667
```

```python
3. Combinatorics Problems
```

```python
Problem 3.1: Committee Formation
```

```python
Q: From 8 men and 7 women, how many ways can we form a 5-person committee with exactly 3 women?
```

```python
Solution:

Choose 3 women from 7: C(7,3) = 35
Choose 2 men from 8: C(8,2) = 28
Total combinations: 35 √ó 28 = 980
```

```python
Python Verification:
```

```python
from math import comb

women = comb(7, 3)  # 35
men = comb(8, 2)    # 28
total = women * men  # 980
print(f"Total committees: {total}")
```

```python
Problem 4.1: Binomial Distribution
```

```python
Q: In 10 coin tosses, what's the probability of getting exactly 6 heads?
```

```python
Solution:
```

```python
P(X=6) = C(10,6) √ó (0.5)^6 √ó (0.5)^4 = 210 √ó 0.015625 ‚âà 0.2051
```

```python
Python Calculation:
```

```python
from scipy.stats import binom

n, p = 10, 0.5
prob = binom.pmf(6, n, p)
print(f"Probability: {prob:.4f}")  # 0.2051
```

```python
Problem 4.2: Poisson Distribution
```

```python
Q: If calls arrive at 2 per minute, what's the probability of 5 calls in 2 minutes?
```

```python
Solution:
Œª = 2 √ó 2 = 4 (for 2 minutes)
P(X=5) = (e‚Åª‚Å¥ √ó 4‚Åµ)/5! ‚âà 0.1563
```

```python
Python Code:
```

```python
from scipy.stats import poisson

lambda_ = 4  # 2 calls/min √ó 2 minutes
prob = poisson.pmf(5, lambda_)
print(f"Probability: {prob:.4f}")  # 0.1563
```

```python
Problem 5.2: Wait Time Estimation
```

```python
Scenario:
```

```python
Average wait time is 15 minutes
Exponentially distributed
```

```python
Q: Probability of waiting <10 minutes?
```

```python
Solution:
```

```python
from scipy.stats import expon

lambda_ = 1/15  # Rate parameter
prob = expon.cdf(10, scale=15)  # scale = 1/lambda
print(f"Probability: {prob:.4f}")  # 0.4866
```
