```python
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
with open("example.txt", "r") as file:
    for line in file:
        print(line.strip())  # strip() removes extra newlines
```

```python
with open("example.txt", "r") as file:
    lines = file.readlines()
    print(lines)  # List where each element is a line
```

```python
# Example with r+ mode (file must exist)
with open("example.txt", "r+") as file:
    # Read first
    content = file.read()
    print("Original content:", content)
    
    # Write at current position (end of file after read)
    file.write("\nNew line added with r+")
    
    # Move to beginning to read again
    file.seek(0)
    print("Updated content:", file.read())
```

```python
with open("new_file.txt", "w") as file:
    file.write("Hello World!\n")
    file.write("This is Python file handling.")
```

```python
with open("existing.txt", "a") as file:
    file.write("\nAdding new content at the end.")
```

```python
# Example with w+ mode (creates file if doesn't exist)
with open("example_w.txt", "w+") as file:
    # Write first (this truncates the file if it exists)
    file.write("Initial content with w+")
    
    # Move to beginning to read
    file.seek(0)
    print("Content after writing:", file.read())
    
    # Write more content
    file.write("\nAdditional line")
    
    # Move to beginning again to read
    file.seek(0)
    print("Final content:", file.read())
```

```python
with open("source.txt", "r") as source:
    with open("destination.txt", "w") as dest:
        dest.write(source.read())
```

```python
with open("data.txt", "r") as file:
    line_count = len(file.readlines())
    print(f"Total lines: {line_count}")
```

```python
with open("grades.csv", "r") as file:
    for line in file:
        name, score = line.strip().split(",")
        print(f"{name}: {score}")
```
