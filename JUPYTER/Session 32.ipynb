```python
ANOVA (Analysis of Variance)
```

```python
What it is: ANOVA is a statistical test used to compare means among three or more groups to determine if at least one group mean is significantly different from the others.

Key concepts:

Null hypothesis: All group means are equal

Alternative hypothesis: At least one group mean is different

F-ratio: The test statistic (between-group variability / within-group variability)
```

```python
When to use it:

When comparing means of 3+ groups

When your data meets assumptions (normality, homogeneity of variance, independence)

Example: Comparing test scores of students taught with three different teaching methods.
```

```python
1. ANOVA (Analysis of Variance) Problem
```

```python
Scenario: A teacher wants to compare the effectiveness of three teaching methods (A, B, C) on students' test scores.
```

```python
Data:
```

```python
Method A	Method B	Method C
78	          85	     90
82	          88	     92
75	          82	     88
80	          86	     95
```

```python
Question: Is there a statistically significant difference in the mean scores of the three teaching methods?
```

```python
State Hypotheses:

Null (H₀): μA = μB = μC (no difference)

Alternative (H₁): At least one mean is different

Calculate Group Means & Overall Mean (Grand Mean, GM)

Mean A = (78 + 82 + 75 + 80)/4 = 78.75

Mean B = (85 + 88 + 82 + 86)/4 = 85.25

Mean C = (90 + 92 + 88 + 95)/4 = 91.25

Grand Mean (GM) = (78.75 + 85.25 + 91.25)/3 ≈ 85.08

Calculate Sum of Squares (SS)

SS Between (SSB) = Σnᵢ(Meanᵢ - GM)²
= 4(78.75 - 85.08)² + 4(85.25 - 85.08)² + 4(91.25 - 85.08)² ≈ 312.5

SS Within (SSW) = Σ(each value - its group mean)²
= (78-78.75)² + (82-78.75)² + ... + (95-91.25)² ≈ 59.5

Total SS (SST) = SSB + SSW ≈ 372

Degrees of Freedom (df)

df between = k - 1 = 3 - 1 = 2

df within = N - k = 12 - 3 = 9

Mean Squares (MS)

MSB = SSB / df between = 312.5 / 2 ≈ 156.25

MSW = SSW / df within = 59.5 / 9 ≈ 6.61

F-Ratio

F = MSB / MSW ≈ 156.25 / 6.61 ≈ 23.63

Compare with Critical F (from F-table at α = 0.05, df = 2,9)

Critical F ≈ 4.26

Since 23.63 > 4.26, we reject H₀.
```

```python
Conclusion: There is a significant difference in test scores among the teaching methods.
```

```python
Correlation
```

```python
What it is: Measures the strength and direction of the linear relationship between two continuous variables.

Key concepts:

Range: -1 to +1

+1 = perfect positive correlation

-1 = perfect negative correlation

0 = no linear correlation

Pearson's r: Most common correlation coefficient
```

```python
When to use it:

When examining if/how two variables change together

Before performing regression analysis

Example: Relationship between hours studied and exam scores.
```

```python
2. Correlation Problem
```

```python
Scenario: A researcher wants to check if there’s a relationship between study hours and exam scores.
```

```python
Data:
```

```python
Hours (X)	Score (Y)
2	          60
4	          75
6	          85
8	          95
```

```python
Solution (Pearson’s r)
```

```python
Calculate Means:

Mean X = (2 + 4 + 6 + 8)/4 = 5

Mean Y = (60 + 75 + 85 + 95)/4 = 78.75

Compute Cov(X,Y), SD(X), SD(Y)

Cov(X,Y) = Σ[(Xᵢ - X̄)(Yᵢ - Ȳ)] / (n-1)
= [(2-5)(60-78.75) + (4-5)(75-78.75) + ...] / 3 ≈ 23.33

SD(X) = √[Σ(Xᵢ - X̄)² / (n-1)] ≈ 2.58

SD(Y) = √[Σ(Yᵢ - Ȳ)² / (n-1)] ≈ 15.48

Pearson’s r = Cov(X,Y) / (SD(X) × SD(Y)) ≈ 23.33 / (2.58 × 15.48) ≈ 0.99
```

```python
Conclusion: There’s a very strong positive correlation (r ≈ 0.99) between study hours and exam scores.
```

```python
Covariance
```

```python
What it is: Measures how much two random variables vary together (direction of linear relationship).

Key concepts:

Positive value = variables tend to move in same direction

Negative value = variables tend to move in opposite directions

Unlike correlation, covariance is not standardized (hard to interpret magnitude)
```

```python
When to use it:

Understanding directional relationship between variables

Building blocks for other statistical measures (like correlation)

Example: Relationship between temperature and ice cream sales.
```

```python
Key Differences
```

```python
Purpose:

ANOVA compares group means

Correlation measures relationship strength/direction

Covariance measures joint variability

Scale:

ANOVA uses F-statistic

Correlation ranges from -1 to 1

Covariance has no fixed range

Variables:

ANOVA: 1 categorical + 1 continuous variable

Correlation/Covariance: 2 continuous variables
```

```python
3. Covariance Problem
```

```python
Scenario: An investor wants to see if stock prices of Company A and B move together.
```

```python
Data:
```

```python
Day	Stock A (X)	Stock B (Y)
1	  50	       80
2	  52	       82
3	  54	       85
4	  53	       84
```

```python
Solution
```

```python
Calculate Means:

Mean X = (50 + 52 + 54 + 53)/4 = 52.25

Mean Y = (80 + 82 + 85 + 84)/4 = 82.75

Compute Cov(X,Y) = Σ[(Xᵢ - X̄)(Yᵢ - Ȳ)] / (n-1)
= [(50-52.25)(80-82.75) + (52-52.25)(82-82.75) + ...] / 3
≈ 3.58
```

```python
Interpretation:
```

```python
Positive covariance (3.58) means Stock A and B tend to move in the same direction.

However, unlike correlation, covariance does not tell us the strength of the relationship.
```

```python
Practical Implementation of ANOVA Test
```

```python
ANOVA (Analysis of Variance) is used to compare means across three or more groups to determine if at least one group is significantly different.
```

```python
When to Use ANOVA?
✅ Compare means of 3+ independent groups (e.g., different teaching methods, drug doses, marketing strategies).
✅ Works for continuous data (e.g., test scores, sales figures).
✅ Assumptions:

Normally distributed data (check with Shapiro-Wilk test).

Homogeneity of variance (equal variances, checked with Levene’s test).

Independent samples.

```

```python
Example Problem: Comparing Exam Scores Across 3 Teaching Methods
```

```python
Scenario: A school tests three teaching methods (A, B, C) to see which works best.
```

```python
Data:
```

```python
Method A	Method B	Method C
78	          85	      90
82	          88	      92
75	          82	      88
80	          86	      95
```

```python
Question: Is there a statistically significant difference in exam scores between the three methods?
```

```python
Step 1: Check Assumptions
1. Normality Check (Shapiro-Wilk Test)
H₀: Data is normally distributed.

H₁: Data is not normally distributed.

If p > 0.05, normality holds.
```

```python
import scipy.stats as stats

# Data for each method
method_A = [78, 82, 75, 80]
method_B = [85, 88, 82, 86]
method_C = [90, 92, 88, 95]

# Shapiro-Wilk test for normality
print(stats.shapiro(method_A).pvalue)  # p > 0.05?
print(stats.shapiro(method_B).pvalue)  # p > 0.05?
print(stats.shapiro(method_C).pvalue)  # p > 0.05?
```

```python
Output:

If all p-values > 0.05 → normality assumption is satisfied.
```

```python
2. Homogeneity of Variance (Levene’s Test)
H₀: Variances are equal across groups.

H₁: Variances are unequal.

If p > 0.05, variances are equal.
```

```python
print(stats.levene(method_A, method_B, method_C).pvalue)
```

```python
Output:

If p > 0.05 → equal variances (homoscedasticity holds).
```

```python
Step 2: Perform One-Way ANOVA
```

```python
Hypotheses:

H₀: μA = μB = μC (no difference in means)

H₁: At least one mean is different
```

```python
f_stat, p_value = stats.f_oneway(method_A, method_B, method_C)
print("F-statistic =", f_stat)
print("p-value =", p_value)
```

```python
Interpretation:

p-value (0.0002) < 0.05 → Reject H₀.

Conclusion: At least one teaching method leads to significantly different exam scores.
```

```python
Step 3: Post-Hoc Test (Tukey’s HSD)
```

```python
Since ANOVA only tells us that at least one group is different, we use Tukey’s HSD to find which pairs are different.
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import numpy as np

# Combine data into one array
data = np.array(method_A + method_B + method_C)

# Create group labels
groups = ['A']*4 + ['B']*4 + ['C']*4

# Perform Tukey's HSD
tukey = pairwise_tukeyhsd(data, groups, alpha=0.05)
print(tukey.summary())
```

```python
Interpretation:

A vs. C has the largest mean difference (12.5) and is highly significant (p=0.001).

A vs. B and B vs. C are also significant (p < 0.05).
```

```python
Step 4: Visualize Results
```

```python
Boxplot to compare distributions:
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x=groups, y=data)
plt.title("Exam Scores by Teaching Method")
plt.show()
```

```python
Expected Output:

Boxplot shows Method C has the highest median score.
```

```python
Key Takeaways
ANOVA tests if 3+ group means are equal.

Check assumptions (normality, equal variance) before running ANOVA.

If ANOVA is significant (p < 0.05), run Tukey’s HSD to find which groups differ.

Visualize with boxplots for better interpretation.
```
