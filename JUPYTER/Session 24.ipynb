```python
1. Descriptive Statistics
```

```python
Purpose: Summarize and describe the main features of a dataset.
```

```python
Key Measures
```

```python
Measure      Description	         Python Implementation

Central Tendency	
Mean	   Average value	         np.mean(data)
Median	   Middle value	             np.median(data)
Mode	   Most frequent value 	     stats.mode(data)

Dispersion	
Range	  Max - Min	                 np.ptp(data)
Variance  Average squared deviation	 np.var(data)
Std Dev   Square root of variance	 np.std(data)
IQR	      Q3 - Q1	                 stats.iqr(data)

Shape	
Skewness  Symmetry measure	        stats.skew(data)
Kurtosis  Tailedness measure	    stats.kurtosis(data)
```

```python
Example:
```

```python
import numpy as np
from scipy import stats

data = [23, 45, 67, 34, 89, 56, 32]

print(f"Mean: {np.mean(data):.2f}")
print(f"Median: {np.median(data)}")
print(f"Std Dev: {np.std(data):.2f}")
print(f"Skewness: {stats.skew(data):.2f}")
```

```python
Visualization Tools
Histogram: plt.hist(data)

Boxplot: sns.boxplot(data)

Scatterplot: plt.scatter(x,y)
```

```python
2. Inferential Statistics
```

```python
Purpose: Make predictions or inferences about a population based on sample data
```

```python
Key Concepts
```

```python
Concept         	      Description	                              Example
Hypothesis Testing	  Test assumptions about population parameters	T-tests, ANOVA
Confidence Intervals  Range containing population parameter	        95% CI for mean
Regression Analysis	  Relationship between variables	            Linear regression
Effect Size           Magnitude of difference/relationship	        Cohen's d, R²
```

```python
Common Tests
```

```python
Test	        Purpose	                 Python Function

T-tests	        Compare means
One-sample	    Compare to known value	ttest_1samp()
Independent	    Compare 2 groups	    ttest_ind()
Paired	        Compare pre-post	    ttest_rel()
ANOVA	        Compare >2 groups	    f_oneway()

Chi-square	    Categorical relationships
Goodness-of-fit	Compare distributions	chisquare()
Independence	Test associations	    chi2_contingency()
```

```python
Example (T-test):
```

```python
from scipy.stats import ttest_ind
group1 = [23, 45, 34, 56]
group2 = [67, 89, 72, 85]

t_stat, p_value = ttest_ind(group1, group2)
print(f"p-value: {p_value:.4f}")
```

```python
Measures of Central Tendency, Shape & Dispersion
```

```python
1. Measures of Central Tendency
These identify the center or typical value of a dataset.
```

```python
Measure	Description	When to Use	Python Implementation	Example
Mean	Average value	Symmetric, continuous data	np.mean(data)	Average salary
Median	Middle value	Skewed data, outliers	np.median(data)	Housing prices
Mode	Most frequent value	Categorical data	stats.mode(data)[0]	Most common shoe size
```

```python
Example:
```

```python
import numpy as np
from scipy import stats

data = [15, 18, 22, 22, 23, 25, 28, 30]

print(f"Mean: {np.mean(data):.2f}")   # 22.88
print(f"Median: {np.median(data)}")   # 22.5
print(f"Mode: {stats.mode(data)[0]}") # 22
```

```python
2. Measures of Shape
These describe the distribution's symmetry and peakness.
```

```python
Measure	Description	Interpretation	Python Implementation
Skewness	Degree of asymmetry	>0: Right-skewed
<0: Left-skewed
=0: Symmetric	stats.skew(data)
Kurtosis	Tailedness/peakness	>3: Leptokurtic (heavy tails)
<3: Platykurtic (light tails)
=3: Normal	stats.kurtosis(data, fisher=False)
```

```python
Example:
```

```python
print(f"Skewness: {stats.skew(data):.2f}")       # 0.36 (mild right skew)
print(f"Kurtosis: {stats.kurtosis(data):.2f}")   # -0.92 (flatter than normal)
```

```python
3. Measures of Dispersion
These quantify data spread/variability.
```

```python
Measure	Description	Python Implementation	Use Case
Range	Max - Min	np.ptp(data)	Quick spread estimate
Variance	Average squared deviations	np.var(data)	Statistical modeling
Std Deviation	√Variance	np.std(data)	Most common spread measure
IQR	Q3 - Q1 (middle 50%)	stats.iqr(data)	Robust to outliers
MAD	Median absolute deviation	stats.median_abs_deviation(data)	Robust alternative
```

```python
Example:
```

```python
print(f"Range: {np.ptp(data)}")                 # 15 (30-15)
print(f"Variance: {np.var(data):.2f}")          # 23.36
print(f"Std Dev: {np.std(data):.2f}")           # 4.83
print(f"IQR: {stats.iqr(data)}")                # 6.5 (Q3-Q1)
print(f"MAD: {stats.median_abs_deviation(data):.2f}") # 4.45
```

```python
4. Practical Considerations
When to Use Which Measure
```

```python
Scenario	Recommended Measures
Normal distribution	Mean + Std Dev
Skewed data	Median + IQR
Outliers present	Median + MAD
Categorical data	Mode + Frequency
```

```python
Visualization Pairings
Boxplot: Median + IQR

Histogram: Mean + Std Dev

Violin Plot: Shows all measures simultaneously
```

```python
Complete Analysis Example
```

```python
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter

# Load dataset
tips = sns.load_dataset('tips')
bill_data = tips['total_bill'].values  # Convert to numpy array for consistency

print("=== Central Tendency ===")
print(f"Mean: ${np.mean(bill_data):.2f}")
print(f"Median: ${np.median(bill_data):.2f}")

# Most robust way to get mode
def get_mode(data):
    try:
        # Try scipy's mode first
        mode_result = stats.mode(data)
        if hasattr(mode_result, 'mode'):  # Newer scipy versions
            return mode_result.mode[0], mode_result.count[0]
        else:  # Older scipy versions
            return mode_result[0][0], mode_result[1][0]
    except:
        # Fallback to collections.Counter
        counts = Counter(data)
        max_count = max(counts.values())
        modes = [k for k,v in counts.items() if v == max_count]
        return modes[0], max_count if len(modes) == 1 else (modes, max_count)

mode_value, mode_count = get_mode(bill_data)
if isinstance(mode_value, list):
    print(f"Multiple modes: {[f'${m:.2f}' for m in mode_value]} (each appears {mode_count} times)")
else:
    print(f"Mode: ${mode_value:.2f} (appears {mode_count} times)")

print("\n=== Shape ===")
print(f"Skewness: {stats.skew(bill_data):.2f}")
print(f"Kurtosis: {stats.kurtosis(bill_data):.2f}")

print("\n=== Dispersion ===")
print(f"Range: ${np.ptp(bill_data):.2f}")
print(f"Std Dev: ${np.std(bill_data):.2f}")
print(f"IQR: ${stats.iqr(bill_data):.2f}")

# Visualization
sns.histplot(bill_data, kde=True)
plt.axvline(np.mean(bill_data), color='r', linestyle='--', label='Mean')
plt.axvline(np.median(bill_data), color='g', linestyle=':', label='Median')
plt.legend()
plt.show()
```

```python
Problem 1.1: Central Tendency

Dataset: [12, 15, 18, 22, 22, 23, 25, 28, 30, 35]

Tasks:
Calculate mean, median, and mode
Identify which measure best represents the data
```

```python
import numpy as np
from scipy import stats

data = [12, 15, 18, 22, 22, 23, 25, 28, 30, 35]

print(f"Mean: {mean:.2f}")       # 23.00
print(f"Median: {median}")       # 22.5

# Version 1: Modern SciPy (returns ModeResult object)
try:
    mode_result = stats.mode(data)
    if hasattr(mode_result, 'mode'):
        mode_value = mode_result.mode[0] if isinstance(mode_result.mode, np.ndarray) else mode_result.mode
        print(f"Mode (Modern SciPy): {mode_value}")
    else:
        # Version 2: Older SciPy (returns tuple)
        mode_value = mode_result[0][0] if isinstance(mode_result[0], (list, np.ndarray)) else mode_result[0]
        print(f"Mode (Older SciPy): {mode_value}")
except:
    # Version 3: Fallback using collections
    from collections import Counter
    counts = Counter(data)
    mode_value = max(counts.items(), key=lambda x: x[1])[0]
    print(f"Mode (Fallback): {mode_value}")
```

```python
Analysis:
The median (22.5) best represents this slightly right-skewed data as it's less affected by the higher values (30, 35) than the mean.
```

```python
Practical Implementation of Descriptive Statistics
```

```python
1. Central Tendency Measures
```

```python
Mean, Median, and Mode
```

```python
import numpy as np
from scipy import stats
from collections import Counter

data = [12, 15, 18, 22, 22, 23, 25, 28, 30, 35]

# Mean (average)
mean = np.mean(data)
print(f"Mean: {mean:.2f}")  # 23.00

# Median (middle value)
median = np.median(data)
print(f"Median: {median}")  # 22.5

# Mode (most frequent) - robust calculation
counts = Counter(data)
mode_value = max(counts, key=counts.get)
mode_count = counts[mode_value]
print(f"Mode: {mode_value} (appears {mode_count} times)")  # 22 (appears 2 times)
```

```python
When to Use Each:
Mean: Best for normally distributed data without outliers
Median: Better for skewed data or with outliers
Mode: For categorical data or identifying most common values
```

```python
2. Dispersion Measures
```

```python
Range, Variance, Standard Deviation
```

```python
# Range (max - min)
data_range = np.ptp(data)  # "peak to peak"
print(f"Range: {data_range}")  # 23 (35-12)

# Variance
variance = np.var(data)
print(f"Population Variance: {variance:.2f}")  # 50.20

# Sample Variance (with Bessel's correction)
sample_variance = np.var(data, ddof=1)
print(f"Sample Variance: {sample_variance:.2f}")  # 55.78

# Standard Deviation
std_dev = np.std(data)
print(f"Population Std Dev: {std_dev:.2f}")  # 7.09
```

```python
Interquartile Range (IQR)
```

```python
q1 = np.percentile(data, 25)  # 19.5
q3 = np.percentile(data, 75)  # 27.5
iqr = q3 - q1
print(f"IQR: {iqr}")  # 8.0

# Tukey's fences for outlier detection
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr
outliers = [x for x in data if x < lower_fence or x > upper_fence]
print(f"Outliers: {outliers}")  # []
```

```python
3. Shape Measures
```

```python
Skewness and Kurtosis
```

```python
# Skewness (measure of asymmetry)
skewness = stats.skew(data)
print(f"Skewness: {skewness:.2f}")  # 0.36 (positive = right-skewed)

# Kurtosis (measure of tail heaviness)
kurtosis = stats.kurtosis(data)
print(f"Kurtosis: {kurtosis:.2f}")  # -0.92 (negative = lighter tails than normal)
```
