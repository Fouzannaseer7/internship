```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, poisson, bernoulli
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")
```

```python
mu = 50
sigma = 10
prob = norm.cdf(60, mu, sigma)
print("P(X < 60):", round(prob, 4))
```

```python
x = np.linspace(20, 80, 100)
y = norm.pdf(x, mu, sigma)

plt.plot(x, y, label="Normal PDF")
plt.title("Normal Distribution (Œº=50, œÉ=10)")
plt.xlabel("X")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
prob = uniform.cdf(6, 0, 10) - uniform.cdf(4, 0, 10)
print("P(4 ‚â§ X ‚â§ 6):", round(prob, 2))
```

```python
x = np.linspace(0, 10, 100)
y = uniform.pdf(x, loc=0, scale=10)

plt.plot(x, y, label="Uniform PDF", color="orange")
plt.title("Uniform Distribution U(0,10)")
plt.xlabel("X")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
Œª = 3
prob = poisson.pmf(5, Œª)
print("P(X = 5):", round(prob, 4))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define the rate (lambda)
lam = 3

# Generate values and their probabilities
x = np.arange(0, 15)
y = poisson.pmf(x, mu=lam)

# Plot
plt.bar(x, y, color='green')
plt.title("Poisson Distribution (Œª=3)")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

```python
p = 0.7
print("P(X = 1):", bernoulli.pmf(1, p))
print("P(X = 0):", bernoulli.pmf(0, p))
```

```python
x = [0, 1]
y = bernoulli.pmf(x, p)

plt.bar(x, y, color='purple')
plt.title("Bernoulli Distribution (p=0.7)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.xticks([0, 1])
plt.show()
```

```python
sample_means = []

for _ in range(1000):
    sample = np.random.uniform(0, 10, 30)
    sample_means.append(np.mean(sample))

# üìà Graph:
sns.histplot(sample_means, kde=True, color='teal')
plt.title("Central Limit Theorem (Means of 1000 Samples)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```
