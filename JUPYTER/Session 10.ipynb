```python
# Simple if statement
age = 18
if age >= 18:
    print("You are eligible to vote")
```

```python
# If-else structure
temperature = 25
if temperature > 30:
    print("It's hot outside")
else:
    print("It's not too hot")
```

```python
num = float(input("Enter a number: "))

if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")

if num.is_integer():
    print("It's an integer")
else:
    print("It's a float")
```
