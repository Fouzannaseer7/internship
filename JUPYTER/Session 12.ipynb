```python
# Grade classifier
score = 85

if score >= 90:
    print("A")
elif score >= 80:  # Only checked if first condition fails
    print("B")  # â† This executes
elif score >= 70:
    print("C")
else:
    print("Fail")
```

```python
# Basic for loop
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit.upper())
# Output: APPLE BANANA CHERRY

# With range()
for i in range(3):  # 0, 1, 2
    print(f"Loop {i}")
```

```python
# Countdown
count = 5
while count > 0:
    print(count)
    count -= 1  # Don't forget to update!
print("Blast off!")

# Input validation
while True:
    age = input("Enter your age: ")
    if age.isdigit():
        break  # Exit loop if valid number
    print("Invalid input! Try again.")
```

```python
# Find first even number
numbers = [1, 3, 5, 8, 7, 9]
for num in numbers:
    if num % 2 == 0:
        print(f"First even: {num}")
        break  # Stops after finding 8
```

```python
for i in range(10):
    if i == 5:
        break  # Loop stops completely at 5
    print(i)
# Output: 0 1 2 3 4
```

```python
for num in range(10):
    if num % 2 == 0:
        continue  # Skip even numbers
    print(num)
# Output: 1 3 5 7 9
```

```python
# Print odd numbers only
for i in range(10):
    if i % 2 == 0:
        continue  # Skip evens
    print(i)  # Prints 1, 3, 5, 7, 9
```
