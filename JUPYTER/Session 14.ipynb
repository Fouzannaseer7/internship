```python
def greet():
    """This function greets the user"""
    print("Hello! Welcome to Python.")
    
# Calling the function
greet()
```

```python
def describe_pet(name, animal="dog"):
    print(f"I have a {animal} named {name}")

describe_pet("Buddy")  # Uses default animal
describe_pet("Whiskers", "cat")
```

```python
def add_numbers(a, b):
    return a + b

result = add_numbers(5, 3)
print(result)  # Output: 8
```

```python
def calculate(a, b):
    sum = a + b
    difference = a - b
    return sum, difference

results = calculate(10, 5)
print(results)  # Output: (15, 5)
```

```python
def no_return():
    print("This function doesn't return anything")

x = no_return()
print(x)  # Output: None
```

```python
def make_pizza(*toppings):
    print("Making pizza with:")
    for topping in toppings:
        print(f"- {topping}")

make_pizza("pepperoni")
make_pizza("mushrooms", "olives", "onions")
```

```python
def build_profile(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

build_profile(name="Alice", age=25, city="New York")
```

```python
def my_function():
    x = 10  # Local variable
    print(x)

my_function()
print(x)  # Error: x is not defined outside
```

```python
x = 5  # Global variable

def change_x():
    global x
    x = 10

change_x()
print(x)  # Output: 10
```

```python
def greet(name):
    """This function greets the person passed in as a parameter"""
    print(f"Hello, {name}! Welcome to Python programming.")

# Calling the function
greet("Alice")
greet("Bob")
```

```python
def describe_pet(pet_name, animal_type="dog"):
    """Display information about a pet"""
    print(f"\nI have a {animal_type}.")
    print(f"My {animal_type}'s name is {pet_name.title()}.")

# Positional argument
describe_pet("willie")

# Multiple positional arguments
describe_pet("harry", "hamster")

# Keyword arguments
describe_pet(pet_name="molly", animal_type="cat")
describe_pet(animal_type="bird", pet_name="tweety")
```

```python
def calculate_rectangle_area(length, width):
    """Calculate and return the area of a rectangle"""
    area = length * width
    return area

# Using the return value
room_area = calculate_rectangle_area(21,10)
print(f"The room area is {room_area} square feet.")
```
