```python
What is Hadoop?
Apache Hadoop is an open-source framework that allows storing and processing large datasets in a distributed computing environment using clusters of 
computers.
```

```python
 Key Features of Hadoop
Open-source
Distributed storage & processing
Fault-tolerant
Scalable
Handles structured, semi-structured, and unstructured data
```

```python
Core Components of Hadoop
```

```python
1. HDFS (Hadoop Distributed File System)
Storage system of Hadoop
Splits files into blocks (default: 128MB or 256MB)
Stores each block on multiple nodes (replication, default: 3)
Master node: NameNode
Slave nodes: DataNodes

2. MapReduce
Programming model for processing big data
Works in two phases:
Map – Processes input data and generates key-value pairs
Reduce – Aggregates intermediate results from the Map phase

3. YARN (Yet Another Resource Negotiator)
Manages cluster resources
Schedules jobs across nodes
Components:
ResourceManager (global)
NodeManager (per node)

ApplicationMaster (per application)
```

```python
Hadoop Ecosystem Tools
```

```python
| Tool          | Description                                     |
| ------------- | ----------------------------------------------- |
| Hive      | SQL-like interface for querying big data        |
| Pig       | Scripting language for analyzing large datasets |
| Sqoop     | Transfers data between Hadoop and RDBMS         |
| Flume     | Collects log/data streams into HDFS             |
| Oozie     | Job scheduler for Hadoop                        |
| Zookeeper| Coordination service for distributed apps       |
| HBase     | NoSQL database for random real-time access      |
```

```python
HDFS Architecture
NameNode – Stores metadata (file names, block locations)
DataNode – Stores actual data blocks
Block – Unit of data storage (e.g., 128MB)
Replication – Blocks are replicated to provide fault tolerance
```

```python
Advantages of Hadoop
Cost-effective (runs on commodity hardware)
Highly scalable
Fault-tolerant
Flexible with various data types
Fast for processing large data
```

```python
Disadvantages
Not suitable for small datasets
High latency
Complex to manage for beginners
Security is basic by default
```

```python
Real-world Use Cases
Web indexing (e.g., Google, Yahoo)
Log processing
Fraud detection
Recommendation engines (e.g., Netflix, Amazon)
Sentiment analysis from social media
```
