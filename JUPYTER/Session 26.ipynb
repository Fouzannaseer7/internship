```python
from scipy.stats import norm

# Define mean and standard deviation
mean = 0
std_dev = 1

# PDF at x = 1
x = 1
pdf_value = norm.pdf(x, mean, std_dev)

print("ðŸ“Œ PDF Example:")
print(f"PDF at x = {x} for N(0,1): {pdf_value:.4f}")
```

```python
from scipy.stats import norm

# CDF at x = 1
x = 1
cdf_value = norm.cdf(x, mean, std_dev)

print("\nðŸ“Œ CDF Example:")
print(f"CDF at x = {x} for N(0,1): {cdf_value:.4f}")
```

```python
#We calculate statistics using the entire dataset.
import numpy as np

# Define the full population: numbers 1 to 100
population = np.arange(1, 101)

# Population statistics
pop_mean = np.mean(population)
pop_std = np.std(population)  # ddof=0 by default for population

print("ðŸ“Š Population Example")
print("Population Data:", population)
print("Population Mean:", pop_mean)
print("Population Standard Deviation:", round(pop_std, 2))
```

```python
#We take a random subset of the population and calculate statistics.
import random
import numpy as np

# Suppose the same population
population = list(range(1, 101))

# Take a sample of 10 values
sample = random.sample(population, 10)

# Sample statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # ddof=1 for sample standard deviation

print("\nðŸŽ¯ Sample Example")
print("Sample Data:", sample)
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", round(sample_std, 2))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Full population: numbers 1 to 100
population = np.arange(1, 101)

# Population statistics
pop_mean = np.mean(population)
pop_std = np.std(population)

# Plot the population
plt.figure(figsize=(12, 6))
plt.hist(population, bins=20, color='skyblue', alpha=0.6, label='Population')
plt.axvline(pop_mean, color='blue', linestyle='--', label=f'Population Mean = {pop_mean:.1f}')

# Take 3 different samples and plot them
for i in range(1, 4):
    sample = random.sample(list(population), 10)
    sample_mean = np.mean(sample)
    plt.axvline(sample_mean, linestyle='--', label=f'Sample {i} Mean = {sample_mean:.1f}')

# Plot settings
plt.title('Population vs Sample Means')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import random
import numpy as np

# Create a population of 100 elements (e.g., 1 to 100)
population = list(range(1, 101))
```

```python
sample_size = 10
simple_random_sample = random.sample(population, sample_size)

print("ðŸŽ² Simple Random Sample:", simple_random_sample)
```

```python
k = 10  # interval
systematic_sample = population[::k]  # every 10th element

print("ðŸ“ Systematic Sample:", systematic_sample)
```

```python
# Split into two strata: even and odd numbers
even_group = [x for x in population if x % 2 == 0]
odd_group = [x for x in population if x % 2 != 0]

# Take 5 from each group
stratified_sample = random.sample(even_group, 5) + random.sample(odd_group, 5)

print("ðŸ“š Stratified Sample:", stratified_sample)
```

```python
# Divide into clusters of 10
clusters = [population[i:i + 10] for i in range(0, len(population), 10)]

# Randomly select 2 clusters
selected_clusters = random.sample(clusters, 2)

# Flatten the result into one sample
cluster_sample = [item for cluster in selected_clusters for item in cluster]

print("ðŸ§© Cluster Sample:", cluster_sample)
```

```python
convenience_sample = population[:10]

print("ðŸª‘ Convenience Sample:", convenience_sample)
```
