 Dictionary Data Type in Python

```python
#Unordered:
my_dict = {"a": 1, "b": 2, "c": 3}
print(my_dict) 
```

```python
#Mutable:
person = {"name": "Ali"}
person["age"] = 25         # Add
person["name"] = "Jabbar"  # Update
del person["age"]          # Delete
print(person)
```

```python
#Keys must be unique:
d = {"a": 1, "b": 2, "a": 3}
print(d) 
```

```python
#Can store different data types:
info = {
    "name": "Sara",
    "marks": [85, 90, 92],
    "details": {"height": 160, "weight": 55}
}
print(info)
```
