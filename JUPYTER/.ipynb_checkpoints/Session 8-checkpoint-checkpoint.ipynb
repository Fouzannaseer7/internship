```python
my_list = [10, "hello", 3.14, True]
print(my_list)
```

```python
# ðŸ”¸ Ordered: Items are stored in the same order
print("Ordered list:", my_list)
```

```python
# ðŸ”¸ Mutable: We can change, add, or remove items
my_list[1] = "banana"      # Changing an element
my_list.append("new item") # Adding an element
my_list.remove(True)       # Removing an element

print("After modifying (Mutable):", my_list)
```

```python
# ðŸ”¸ Allows Duplicates: 'apple' appears more than once
print("Duplicates allowed:", my_list)
```

```python

# ðŸ”¸ Heterogeneous: List contains different data types
print("Mixed types in list:")
for item in my_list:
    print(f"{item} ({type(item)})")
```

```python
my_list = ['a', 'b', 'c', 'd']
print(my_list[0])   
print(my_list[1])  
print(my_list[2])  
print(my_list[3])  
print(my_list[-1])  
print(my_list[4])  
```

```python
my_list = [0, 1, 2, 3, 4, 5]
print(my_list[1:4])    # [1, 2, 3]
print(my_list[:3])     # [0, 1, 2]
print(my_list[::2])    # [0, 2, 4]
```

```python
a = 'python class'
print(a[-6::3])
```

```python
lst = [1, 2]
lst.append(3)
print(lst)  # [1, 2, 3]
```

```python
lst = [1, 3]
lst.insert(1, 2)
print(lst)  # [1, 2, 3]
```

```python
lst = [1, 2]
lst.extend([3, 4])
print(lst)  # [1, 2, 3, 4]
```

```python
lst = [1, 2, 3, 2]
lst.remove(2)
lst.remove(3)
print(lst)  
```

```python
lst = [1, 2, 3]
x = lst.pop()
print(x)     
print(lst)   
```

```python
lst = ['a', 'b', 'c']
print(lst.index('b'))  
```

```python
lst = [1, 2, 2, 3]
print(lst.count(2)) 
print(lst.count(5)) 
```

```python
lst = [3, 1, 2]
lst.sort()
print(lst) 
```

```python
lst = [1, 2, 3]
lst.reverse()
print(lst) 
```

```python
lst1 = [1, 2, 3]
lst2 = lst1.copy()
print(lst2) 
```

```python
lst = [1, 2, 3]
lst.clear()
print(lst) 
```
