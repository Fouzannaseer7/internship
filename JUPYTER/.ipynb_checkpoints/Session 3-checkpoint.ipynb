```python
a = 10
b = 3

print(a + b)  
print(a - b)  
print(a * b) 
print(a / b) 
print(a // b) 
print(a % b) 
print(a ** b) 
```

```python
name = "Ali"
greet = "Hello"

# Concatenation
print(greet + " " + name)   

# Repetition
print(name * 3)             

# Indexing
print(name[0])            

# Slicing
print(name[1:])           

# Length
print(len(name))          
```

```python
fruits = ["apple", "banana"]

# Add item
fruits.append("cherry")

# Remove item
fruits.remove("banana")

# Indexing
print(fruits[0])  

# Slicing
print(fruits[0:2])  

# Length
print(len(fruits))  
```

```python
s1 = {1, 2, 3}
s2 = {3, 4, 5}

print(s1.union(s2))        
print(s1.intersection(s2)) 
print(s1 - s2)             
```

```python
person = {"name": "Ali", "age": 25}

# Access value
print(person["name"])

# Add new key
person["city"] = "Delhi"

# Keys and values
print(person.keys())    
print(person.values())  
```
