```python
a = "hello"
b = "hello"
print(id(a))  # Memory address of a
print(id(b))  # Same as a, because strings are immutable
```

```python
text = "Python"
print(text[0])   
print(text[3])   
print(text[-1])  
print(text[-3])  
```

```python
‚úÖ Rules:
 >Syntax: string[start:stop:step]
  \\start: index to begin (inclusive)
  \\stop: index to end (exclusive)
  \\step: how many characters to skip (default is 1)
 >Omitting start starts from the beginning.
 >Omitting stop goes till the end.
 >Slicing does not cause an error if start or stop is out of bounds ‚Äî it adjusts safely.
 >Negative step can be used to reverse the string.
```

```python
s = "PythonProgramming"

print(s[0:6])    
print(s[6:])      
print(s[:6])      
print(s[::2])    
print(s[::-1])   
```

```python
# MEMORY
print("üîÅ MEMORY EXAMPLE")
a = "hello"
b = "hello"
c = "HELLO"
print("ID of a:", id(a))
print("ID of b:", id(b))
print("ID of c:", id(c))  
print("a and b share memory:", id(a) == id(b))
print("a and c share memory:", id(a) == id(c))

print("\nüî¢ INDEXING EXAMPLE")
text = "Python"
print("Original string:", text)
print("text[0] =", text[0])    
print("text[3] =", text[3])    
print("text[-1] =", text[-1]) 


print("\n‚úÇÔ∏è SLICING EXAMPLE")
msg = "PythonProgramming"
print("Original string:", msg)
print("msg[0:6] =", msg[0:6])       
print("msg[6:] =", msg[6:])         
print("msg[:6] =", msg[:6])         
print("msg[::2] =", msg[::2])       
print("msg[::-1] =", msg[::-1])     
print("msg[5:100] =", msg[5:100])   
```
