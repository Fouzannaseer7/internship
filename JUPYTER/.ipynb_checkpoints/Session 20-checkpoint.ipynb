```python
Matplotlib and Seaborn Visualization
```

```python
>Data visualization is a crucial skill for anyone working with data. 
>Matplotlib and Seaborn are two of the most popular Python libraries for creating static, animated, and interactive visualizations. 
>Here's a beginner-friendly guide to get you started.
```

```python
1.Basic Matplotlib Plots
```

```python
Line Plot
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a figure and axis
plt.figure(figsize=(8, 4))  # (width, height) in inches

# Plot the data
plt.plot(x, y, color='blue', linestyle='-', linewidth=2, marker='o', label='Linear')

# Add title and labels
plt.title('Simple Line Plot', fontsize=14)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

# Add legend
plt.legend()

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
Bar Chart
```

```python
categories = ['A', 'B', 'C', 'D']
values = [15, 25, 30, 20]

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=['red', 'green', 'blue', 'orange'])
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
Scatter Plot
```

```python
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()  # Show color scale
plt.title('Scatter Plot with Color and Size')
plt.show()
```

```python
2.Basic Seaborn Plots:
>Seaborn is built on top of Matplotlib and provides a higher-level interface for 
 statistical graphics.
```

```python
Histogram with Kernel Density Estimate (KDE)
```

```python
import seaborn as sns

# Load example dataset
tips = sns.load_dataset('tips')

# Create plot
sns.histplot(data=tips, x='total_bill', kde=True)
plt.title('Distribution of Total Bill Amounts')
plt.show()
```

```python
Box Plot
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title('Total Bill by Day and Gender')
plt.show()
```

```python
Violin Plot
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)
plt.title('Violin Plot of Total Bill by Day and Gender')
plt.show()
```

```python
Pair Plot (Scatter Matrix)
```

```python
sns.pairplot(tips, hue='sex', diag_kind='kde')
plt.suptitle('Pair Plot of Tips Dataset', y=1.02)
plt.show()
```

```python
3.Customizing Plots
```

```python
Style and Context
```

```python
# Set style (options: darkgrid, whitegrid, dark, white, ticks)
sns.set_style('whitegrid')

# Set context (options: paper, notebook, talk, poster)
sns.set_context('notebook')

plt.figure(figsize=(8, 6))
sns.barplot(data=tips, x='day', y='total_bill', errorbar=None)
plt.title('Average Total Bill by Day')
plt.show()
```

```python
Color Palettes
```

```python
# Set color palette
sns.set_palette('husl')

plt.figure(figsize=(8, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', size='size', sizes=(20, 200))
plt.title('Tips vs Total Bill by Day')
plt.show()
```

```python
4.Combining Matplotlib and Seaborn
```

```python
You can use Seaborn for the statistical plotting and Matplotlib for additional customization:
```

```python
plt.figure(figsize=(10, 6))
ax = sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')

# Matplotlib customization
ax.set_title('Total Bill Distribution by Day and Gender', fontsize=16)
ax.set_xlabel('Day of Week', fontsize=14)
ax.set_ylabel('Total Bill ($)', fontsize=14)
ax.legend(title='Gender', fontsize=12)

plt.show()
```

```python
5.Saving Plots
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Tip vs Total Bill')

# Save as PNG (also supports PDF, SVG, etc.)
plt.savefig('tip_vs_bill.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
1.Histogram (Age Distribution)
```

```python
Matplotlib
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
titanic = sns.load_dataset('titanic')

# Plot
plt.figure(figsize=(8, 5))
plt.hist(titanic['age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution (Matplotlib)')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
Seaborn
```

```python
import seaborn as sns

# Load data
titanic = sns.load_dataset('titanic')

# Plot
sns.histplot(data=titanic, x='age', bins=20, kde=True, color='salmon')
plt.title('Age Distribution with KDE (Seaborn)')
plt.show()
```

```python
2.Bar Chart (Survival Rate by Class)
```

```python
Matplotlib
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load and prepare data
titanic = sns.load_dataset('titanic')
survival_rate = titanic.groupby('class', observed=True)['survived'].mean()

# Plot
plt.figure(figsize=(8, 5))
plt.bar(survival_rate.index, survival_rate.values, color=['red', 'green', 'blue'])
plt.title('Survival Rate by Class (Matplotlib)')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.show()
```

```python
Seaborn
```

```python
import seaborn as sns

# Load data
titanic = sns.load_dataset('titanic')

# Plot
sns.barplot(data=titanic, x='class', y='survived', errorbar=None)
plt.title('Survival Rate by Class (Seaborn)')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.show()
```

```python
3.Box Plot (Age by Survival Status)
```

```python
Matplotlib
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
titanic = sns.load_dataset('titanic')

# Prepare data
died_age = titanic[titanic['survived'] == 0]['age'].dropna()
survived_age = titanic[titanic['survived'] == 1]['age'].dropna()

# Plot with corrected parameter name
plt.figure(figsize=(8, 5))
plt.boxplot([died_age, survived_age], 
            tick_labels=['Died', 'Survived'])  # Changed from 'labels' to 'tick_labels'
plt.title('Age Distribution by Survival Status (Matplotlib)')
plt.ylabel('Age')
plt.show()
```

```python
Seaborn
```

```python
import seaborn as sns

# Load data
titanic = sns.load_dataset('titanic')

# Plot
sns.boxplot(data=titanic, x='survived', y='age')
plt.title('Age Distribution by Survival (Seaborn)')
plt.xticks([0, 1], ['Died', 'Survived'])
plt.show()
```

```python
4. Scatter Plot (Age vs. Fare)
```

```python
Matplotlib
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
titanic = sns.load_dataset('titanic')

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(titanic['age'], titanic['fare'], alpha=0.5, color='purple')
plt.title('Age vs. Fare (Matplotlib)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
Seaborn
```

```python
import seaborn as sns

# Load data
titanic = sns.load_dataset('titanic')

# Plot
sns.scatterplot(data=titanic, x='age', y='fare', alpha=0.5, hue='survived')
plt.title('Age vs. Fare (Colored by Survival) (Seaborn)')
plt.show()
```

```python
5. Count Plot (Passengers by Class & Gender)
```

```python
Matplotlib
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
titanic = sns.load_dataset('titanic')

# Prepare data with observed=True
counts = titanic.groupby(['class', 'sex'], observed=True).size().unstack()

# Plot
counts.plot(kind='bar', stacked=True, color=['pink', 'lightblue'])
plt.title('Passengers by Class & Gender (Matplotlib)')
plt.ylabel('Count')
plt.show()
```

```python
Seaborn
```

```python
import seaborn as sns

# Load data
titanic = sns.load_dataset('titanic')

# Plot
sns.countplot(data=titanic, x='class', hue='sex', palette=['pink', 'lightblue'])
plt.title('Passengers by Class & Gender (Seaborn)')
plt.ylabel('Count')
plt.show()
```
