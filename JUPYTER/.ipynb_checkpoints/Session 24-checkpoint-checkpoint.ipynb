```python
#Step-by-Step Example in Python:
import pandas as pd
import numpy as np
from statistics import mode

# Sample data
data = [70, 85, 90, 75, 88, 90, 76, 95, 90, 80]

# Convert to a pandas Series
scores = pd.Series(data)

# Descriptive statistics
print("Count:", scores.count())
print("Mean:", scores.mean())
print("Median:", scores.median())
print("Mode:", mode(data))
print("Min:", scores.min())
print("Max:", scores.max())
print("Range:", scores.max() - scores.min())
print("Standard Deviation:", scores.std())
print("Variance:", scores.var())
```

```python
import numpy as np
import scipy.stats as stats

# Sample data
data = [68, 75, 80, 70, 72, 74, 78, 79, 82, 76]

# Mean and standard error
mean = np.mean(data)
se = stats.sem(data)  # standard error

# 95% confidence interval
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=se)
print("95% Confidence Interval:", ci)
```

```python
# Is sample mean different from 75?
t_stat, p_value = stats.ttest_1samp(data, 75)
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
group1 = [80, 85, 88, 90, 92]
group2 = [78, 76, 74, 79, 75]

t_stat, p_value = stats.ttest_ind(group1, group2)
print("T-test between two groups:")
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
# Observed frequencies
observed = [20, 30, 50]
expected = [25, 25, 50]

chi2, p = stats.chisquare(f_obs=observed, f_exp=expected)
print("Chi-square:", chi2)
print("P-value:", p)
```

```python
import numpy as np
from scipy import stats

data = [70, 80, 75, 90, 70]
print("Mean:", np.mean(data))        
print("Median:", np.median(data))  
print("Mode:", stats.mode(data))    
```

```python
from scipy.stats import skew, kurtosis

print("Skewness:", skew(data))     
print("Kurtosis:", kurtosis(data)) 
```

```python
print("Range:", max(data) - min(data))     
print("Variance:", np.var(data))           # Spread
print("Standard Deviation:", np.std(data)) # How far from mean
```

```python
# Student Marks Analysis
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data: student marks out of 100
marks = [45, 67, 89, 76, 56, 45, 90, 72, 60, 68, 79, 83, 45, 77, 66, 59, 74, 81, 90, 92]

# Convert to Series (optional)
marks_series = pd.Series(marks)

# --- Measures of Central Tendency ---
print("ðŸ“Œ Central Tendency:")
print("Mean:", np.mean(marks))
print("Median:", np.median(marks))

# Fixing the mode line to work in all scipy versions
mode_result = stats.mode(marks, keepdims=True)
print("Mode:", mode_result.mode[0])

# --- Measures of Dispersion ---
print("\nðŸ“Œ Dispersion:")
print("Range:", max(marks) - min(marks))
print("Standard Deviation:", np.std(marks))
print("Variance:", np.var(marks))

# --- Measures of Shape ---
print("\nðŸ“Œ Shape:")
print("Skewness:", stats.skew(marks))
print("Kurtosis:", stats.kurtosis(marks))

# --- Optional: Visualize the distribution ---
plt.figure(figsize=(10,5))
sns.histplot(marks, kde=True, bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Student Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
