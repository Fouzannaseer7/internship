```python
a = 10           # int
b = 3.14         # float
c = "Hello"      # str
d = True         # bool
e = None         # NoneType
f = 2 + 3j       # complex

print(type(a))   
print(type(b))   
print(type(c))   
print(type(d))   
print(type(e))   
print(type(f))   
```

```python
# list
my_list = [1, 2, 3]
print(type(my_list))  

# tuple
my_tuple = (1, 2, 3)
print(type(my_tuple))

# dict
my_dict = {"name": "Ali", "age": 25}
print(type(my_dict))   

# set
my_set = {1, 2, 2, 3}
print(my_set)         
print(type(my_set))    

# range
r = range(5)
print(list(r))         
print(type(r))         
```
