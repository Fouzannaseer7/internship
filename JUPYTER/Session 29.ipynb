```python
1. Type I and Type II Errors
```

```python
The Analogy:
Imagine a fire alarm:

Type I Error (False Positive): Alarm rings when there’s no fire (you panic unnecessarily).

Type II Error (False Negative): Alarm fails to ring during a fire (you ignore a real danger).

In Statistics:
Error Type	 Null Hypothesis (H₀)	        Reality	                        Consequence
Type I (α)	 Reject H₀ (claim an effect)	H₀ is true (no effect)	    False discovery (e.g., approving a bad drug)
Type II (β)	 Fail to reject H₀ (no effect)	H₀ is false (effect exists)	Missed opportunity (e.g., rejecting a good drug)
```

```python
Example:
H₀: "This new drug has no effect."

Type I: Conclude the drug works when it doesn’t.

Type II: Conclude the drug doesn’t work when it actually does.

Trade-off:
Reducing Type I errors (α) increases Type II errors (β), and vice versa.
```

```python
2. p-values
```

```python
What is a p-value?
The probability of observing your data (or more extreme) if H₀ is true.

Low p-value (≤ 0.05): Unlikely under H₀ → Reject H₀.

High p-value (> 0.05): Likely under H₀ → No evidence to reject H₀.

Example:
Experiment: Test if a coin is fair (H₀: p = 0.5).

Data: Flip 10 times, get 9 heads.

p-value: Probability of getting ≥9 heads if fair = ~0.01.

Conclusion: p < 0.05 → Reject H₀ (coin is likely biased).
```

```python
Common Misinterpretations:
❌ "p = 0.04 means there’s a 4% chance H₀ is true."
✅ Correct: "If H₀ were true, there’s a 4% chance of seeing this result."
```

```python
3. Confidence Intervals (CI)
```

```python
What is a 95% CI?
A range of values that has a 95% probability of containing the true population parameter (e.g., mean, proportion).

Interpretation: "We’re 95% confident the true value lies between X and Y."

Example:
Study: Average height of a sample = 170 cm (95% CI: [165, 175]).

Meaning: If we repeated the study 100 times, ~95 of those CIs would contain the true average height.
```

```python
Key Points:
Wider CI = More uncertainty (small sample size/high variability).

Narrower CI = More precision (large sample size/low variability).

CI vs. p-value:

If a 95% CI excludes the null value (e.g., 0 for mean difference), p < 0.05.
```

```python
Z-Test: One-Tailed vs. Two-Tailed 
```

```python
1. What is a Z-Test?
A Z-test is a statistical test used to determine whether:

A sample mean differs significantly from a population mean (when population standard deviation is known), or

Two sample means differ significantly from each other
```

```python
When to Use It?
✅ Population standard deviation (σ) is known.
✅ Sample size is large (typically n > 30).
✅ Data is normally distributed.
```

```python
2. Types of Z-Tests
```

```python
A) Two-Tailed Z-Test (Non-Directional)
Tests for any difference (≠) between sample and population means.

Null Hypothesis (H₀):

"The sample mean is equal to the population mean (μ = μ₀)."

Alternative Hypothesis (H₁):

"The sample mean is not equal to the population mean (μ ≠ μ₀)."

Critical Region: Split into both tails of the distribution.

Example:
H₀: The average IQ is 100 (μ = 100).

H₁: The average IQ is not 100 (μ ≠ 100).

Test: If sample mean is significantly higher or lower than 100.
```

```python
       Reject H₀          Fail to Reject H₀          Reject H₀
         (α/2)                  (1-α)                 (α/2)
      <---|------------------------|------------------------|--->
          -Zα/2                    0                    +Zα/2
```

```python
B) One-Tailed Z-Test (Directional)
Tests for a difference in one direction only (> or <).

Two Types:

Right-Tailed (Upper-Tail) Test (Tests if μ > μ₀)

Left-Tailed (Lower-Tail) Test (Tests if μ < μ₀)

Example (Right-Tailed):
H₀: The average salary is ≤ $50,000 (μ ≤ 50,000).

H₁: The average salary is > $50,000 (μ > 50,000).

Test: Only checks if the sample mean is significantly higher.
```

```python
5. One-Tailed vs. Two-Tailed Comparison
Feature	           Two-Tailed Test	           One-Tailed Test
Hypothesis	        H₁: μ ≠ μ₀	               H₁: μ > μ₀ or μ < μ₀
Critical Region	    Both tails (α/2 each)	    One tail (full α)
When to Use	        Testing for any difference	Testing for increase/decrease
Critical Z (α=0.05)	±1.96	                    +1.645 (Right) or -1.645 (Left)
```

```python
6. Key Takeaways
✔ Two-tailed tests check for any difference (more conservative).
✔ One-tailed tests check for increase/decrease (more power if direction is known).
✔ Z-test requires known σ and large sample size (else use t-test).
✔ Reject H₀ if Z-score exceeds the critical value.
```
