```python
üîπ Rules:
 >is checks if two variables point to the same memory.
 >Use == to compare content, use is to compare memory address.
 >Python may reuse memory for identical string literals.
```

```python
a = "hello"
b = "hello"
c = "".join(["he", "llo"])  # Created differently

print(a is b)   # True (same literal)
print(a == c)   # True (same content)
print(a is c)   # False (different object in memory)
```

```python
a = "data"
b = "".join(["da", "ta"])
a == b     # ‚úÖ True (same content)
a is b     # ‚ùå False (different memory)
```

```python
# Concatenation
greeting = "Good"
time = "Morning"
print("Concatenation:", greeting + " " + time)

# Immutability
msg = "Hello"
# msg[0] = "M"  # TypeError
new_msg = "M" + msg[1:]
print("Immutable changed:", new_msg)

# Membership
sentence = "I love Python"
print("'love' in sentence:", 'love' in sentence)
print("'hate' not in sentence:", 'hate' not in sentence)

# Identity
a = "data"
b = "data"
c = "".join(["da", "ta"])
print("a == c:", a == c)    # True
print("a is c:", a is c)    # False

# Repetition
print("Echo:", "Hey! " * 3)
```
