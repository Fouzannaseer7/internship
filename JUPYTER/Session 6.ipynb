```python
name = "  Python Programming  "

# Remove extra spaces
cleaned = name.strip()
print(cleaned)  

# Convert to lowercase
print(cleaned.lower())  

# Replace word
print(cleaned.replace("Python", "Java"))  

# Check if it starts with "Py"
print(cleaned.startswith("Py"))  

# Count how many times 'm' appears
print(cleaned.count("m"))  
```
